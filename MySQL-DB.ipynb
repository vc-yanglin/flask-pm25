### 增加約束欄位避免重複

```python
sqlstr='''
create table if not exists pm25(
id int auto_increment primary key,
site varchar(20) not null,
county varchar(50) not null,
pm25 float not null,
datacreationdate datetime not null,
itemunit varchar(10),
unique key uq_site_datetime (site,datacreationdate)
);

'''
print(sqlstr)
```

```python
cur.execute(sqlstr)
conn.commit()
```

### 寫入資料  
- 使用%s佔位跟ignore

```python
sqlstr='''
insert ignore into pm25(site,county,pm25,datacreationdate,itemunit) 
values(%s,%s,%s,%s,%s)
''' 
```

```python
values=df1.values.tolist()
values
```

```python
cur.execute(sqlstr,values[0])
conn.commit()
```

### executemany 一次寫入多筆資料

```python
cur.executemany(sqlstr,values)
conn.commit()
```

### select from 
- group by   (群分)
- order by   (排序)
- asc,desc   (升/降)

```python
sqlstr="""select county,ROUND(AVG(pm25),2) as avg_pm25
from pm25 
group by county 
order by avg_pm25 desc;
"""
cur.execute(sqlstr)
rows=cur.fetchall()
rows
```

### 找出臺北市站點PM2.5平均數據

```python
df1.groupby("county").get_group("臺北市").groupby("site")["pm25"].mean().round(2)
```

```python
sqlstr="""
select site,round(AVG(pm25),2) as avg_pm25
from pm25
where county='臺北市'
group by site
order by avg_pm25 desc
"""

cur.execute(sqlstr)
rows=cur.fetchall()
rows
```
